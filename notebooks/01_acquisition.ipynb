Cette étape correspond au premier chapitre du pipeline de Data Science. L'objectif est d'identifier, d'importer et de consolider vos jeux de données bruts issus de différentes sources (fichiers CSV locaux, requêtes API, bases de données, etc.).

### 1. Initialisation de l'environnement

```python
import os
import sys
import pandas as pd
import numpy as np

# Ajout du dossier parent au chemin de recherche des modules
sys.path.append(os.path.abspath('..'))

print("Libraries importées avec succès ! Prêt pour l'étape d'Acquisition des Données.")
```

### 2. Chargement de la source de données principale

Dans cette première phase d'acquisition des données, nous importons notre jeu de données brut **Automobile** depuis le fichier CSV local `data/raw/Automobile.csv` pour obtenir un premier aperçu complet avant l'étape de nettoyage.

**Objectif de l'étude :** Analyser les caractéristiques techniques des véhicules (taille du moteur, poids, type de carburant, chevaux, etc.) afin de comprendre les facteurs clés qui influencent directement leur prix sur le marché et leurs performances.

```python
# Chemin vers le jeu de données principal
main_data_path = '../data/raw/Automobile.csv'

# Chargement du dataset avec Pandas
df_main = pd.read_csv(main_data_path)

# Affichage de la taille et des premières lignes
print(f"Dimensions du dataset principal : {df_main.shape}")
df_main.head()
```

### 3. Intégration de données secondaires (Multi-Sources)

Création d'un DataFrame complémentaire contenant des informations géographiques sur l'origine des véhicules (région et code continent).

```python
# Création d'un DataFrame secondaire avec les informations géographiques par origine
origin_info = pd.DataFrame({
    'origin': ['usa', 'europe', 'japan'],
    'region': ['Amérique du Nord', 'Europe', 'Asie'],
    'continent_code': ['NA', 'EU', 'AS']
})

print(f"Dimensions des données secondaires : {origin_info.shape}")
print("Données secondaires prêtes :")
origin_info.head()
```

### 4. Fusion des sources (Optionnel)

Jointure du dataset principal avec les données géographiques secondaires sur la colonne `origin`.

```python
# Fusion du dataset principal avec les informations géographiques
df_merged = pd.merge(df_main, origin_info, on='origin', how='left')

print(f"Dimensions après fusion : {df_merged.shape}")
df_merged.head()
```

### 5. Consignation des données d'entrée brutes

Sauvegardez l'état brut de vos données d'entrée pour la suite du pipeline.

```python
# Sauvegarde du dataset fusionné
output_path = '../data/raw/Automobile_merged.csv'
df_merged.to_csv(output_path, index=False)

print(f"Dataset fusionné sauvegardé : {output_path}")
print(f"Dimensions finales : {df_merged.shape}")
print("Acquisition terminée avec succès !")
```
